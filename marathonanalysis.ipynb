## Data Cleaning

### Only Want USA Races, 50k or 50 Mi, 2020

```python
# Step 1: Show both 50k and 50mi
# Note: Distance labeled as 'km' and 'mi'

df[df['Event distance/length'].isin(['50km','50mi'])]
# df[(df['Event distance/length'] == '50km') | (df['Event distance/length'] == '50mi')]
```

```python
# Step 2: Add Races From 2020
# Note: Year of event is an int, not a string

df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)]
```

```python
# Step 3: Add USA Races
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains("(USA)"))]
```

```python
df2.head(10)
```

```python
#Remove '(USA)' From Event Name
df2['Event name'].str.replace(r'\(USA\)','')
```

```python
df2['Event name'] = df2['Event name'].str.replace(r'\(USA\)','')
```

```python
df2.head()
```

```python
#Create Athlete Age Column
#Age = Year of event - Athlete year of birth
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#Remove h From Athlete Performance

df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
#Removing Columns
#Drop: Athlete club, Athlete country, Athlete year of birth, Athlete age category
df2 = df2.drop(columns = ['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'])
```

```python
df2.head()
```

```python
#Clean Up Null Values
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
# Check For Duplicates
df2[df2.duplicated() == True]
```

```python
#Reset Index
df2.reset_index(drop = True)
```

```python
#Fix Types
df2.dtypes
```

```python
#Change athlete_age dtype from float to int
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
#Change Athlete average speed from object to float
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#Rename Columns

df2 = df2.rename(columns = {'Year of event' :'year',
                            'Event dates' : 'race_day',
                            'Event name':'race_name',
                            'Event distance/length':'race_length',
                            'Event number of finishers':'race_number_of_finishers',
                            'Athlete performance':'athlete_performance',
                            'Athlete gender':'athlete_gender',
                            'Athlete average speed':'athlete_average_speed',
                            'Athlete ID':'athlete_id'
                            }
                )
```

```python
df2.head()
```

```python
#Rearrange Column Order 
df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
df3.head()
```

### EDA

```python
#Q1: Difference in speed for the 50k/50mi for male to female
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#Q2: What age groups are the best in the 50m race (20 + races min) (Show 15)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

```python
#Q3: What age groups are the worst in the 50m race (20 + races min) (Show 15)

df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>19').head(15)
```

```python
#Q4: Are racers slower in the summer than winter?

#Adding a race_month column
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
